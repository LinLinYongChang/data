```python
import pandas as pd#导入pandas库
import numpy as np#导入numpy库
```

# 数据输入

```python
n,m=map(int, input("请输入变量数n和约束数m：").split())#定义n，m的输入，要求为整数
values=list(map(float, input("输入%d个物品价值："%n).split()))#定义物品的输入，输入成列表
x_data=[]#定义空列表，后面用来存储输入的属性
for i in range(m):
    print("输入第 %d 个属性："%(i+1),end='')#提示输入第i个属性，end=‘’指去除换行
    t=list(map(int, input().split()))#接收输入的属性，每次输入一行
    x_data.append(t)#将属性添加刚刚定义的x_data列表中
data = pd.DataFrame(x_data)#将属性转换成数据帧的形式
y=list(map(int, input("输入%d个属性的约束条件："%m).split()))#定义接收约束条件的列表
```

# 贪心选择

```python
a_copy=np.array(values.copy())#将价值列表复制一份
for i in range(m):#循环求价值/属性比值
    a_copy=a_copy/(data.iloc[i]+1)
a_values=(pd.Series(a_copy)).sort_values(ascending=False)#将得到的结果按照从大到小排序
a_values_index=a_values.index#拿到排序后的索引
data_1=data[a_values_index]#按照索引改变原来数据帧data顺序
data_2=np.array(data_1)#将数据帧转换成数组
data_3=pd.DataFrame(data_2)#将数组转换成数据帧，这两步会重新建立0—n的列索引
data_3['约束条件']=y#将约束条件加入data数据帧中成为一列
values_series=pd.Series(values)#将原价值转换陈序列
a_array=np.array(values_series.reindex(a_values_index))#将原价值按照之前的序列排序后转换成数组，此步骤为了能将价值和物品一一对应
```

# 回溯法

```python
v=0#定义存放临时价值的初始值
p=0#定义临时变量初始值
pp=0#定义临时变量初始值
bestx=None#存放所取物品的序号，如取第一个物品，第一个位置就为True
x=[False for i in range(n)]#定义临时存放物品序号的列表，初始值为False
sum_11=a_array[0]#初始化最终价值
def f(i):#定义函数
    global v,sum_11,p,pp,x,bestx#声明全局变量
    if i >n-1:#判断是否到叶子节点
        if v>sum_11:#到叶子节点后判断当前价值是否大于最大价值
            sum_11=v#如果大于这把当前价值设为最大价值
            bestx=x[:]#将当前取的物品存放到bestx列表中
    l=np.array(a_array[i:]).sum()#定义一个剪枝函数
    if l+v>sum_11:#满足剪枝函数则继续分支，否则不分支
        p=pp+data_3[i]#将当前i的限制加上已有的限制赋值给p  
        p_1=data_3['约束条件']-p#将约束条件将去限制
        p_2=p_1>=0#判断所有限制是否还大于等于零，大于零为True，小于零为False
        if p_2.all():#判断是否全部约束为1
            x[i]=True#选择当前物品
            pp=pp+data_3[i]#将当前限制加入总限制
            v=v+a_array[i]#将加入物品价值
            f(i+1)#分支
            pp=pp-data_3[i]
            v=v-a_array[i]
        x[i]=False#不取当前物品
        f(i+1)#分支
f(0)#调用函数，i=0开始
gg=[]#定义空列表
#数据输出
for j in range(len(bestx)):
    if bestx[j]==True:
        gg.append(j)
gg=np.array(gg)+1
print("全部物品价值为：",a_array)
print("最大价值取第",gg,"个物品")
print("且最大价值为%f元"%sum_11)
```
